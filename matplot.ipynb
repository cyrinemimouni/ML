```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 2, 10)
y = x**2
print(x)
plt.plot(x, y)
#plt.show()
```

```python
plt.scatter(x, y)
plt.show()
```

```python
plt.plot(x, y, c='red', lw=10, ls='--', alpha=0.5)
plt.show()
```

```python
x = np.linspace(0, 2, 10)
print(x)
plt.figure() # Cr√©ation d'une figure
plt.plot(x, x**2, label='quadratique') # premiere courbe
plt.plot(x, x**3, label='cubique') # deuxieme courbe
# Extra information
plt.title('figure 1') # titre
plt.xlabel('axe x') # axes
plt.ylabel('axe y') # axes
plt.legend() # legend

plt.savefig('figure.png') # sauvegarde la figure dans le repertoire de travail
plt.show() # affiche la figure
```

```python
plt.subplot(2, 1, 1)
plt.plot(x, x*2, c='red')
plt.subplot(2, 1, 2)

plt.plot(x, x**2, c='blue')
```

```python
from numpy.lib.function_base import blackman
fig, ax = plt.subplots(2, 2, sharex=True) # partage le meme axe pour les subplots
ax[0,0].plot(x, y)
ax[0,1].plot(x, np.sin(x),c='red')
ax[1,0].plot(x, np.cos(x), c='green')
ax[1,1].plot(x, np.cos(x))
plt.show()
```

```python
# SOLUTION
def graphique(data):
    n = len(data)
    plt.figure(figsize=(12, 20))
    
    for k, i in zip(data.keys(), range(1, n+1)):
        plt.subplot(n, 1, i)
        plt.plot(data[k])
        plt.title(k)
        
    plt.show()
```
