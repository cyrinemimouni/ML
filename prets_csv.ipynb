```python
import matplotlib.pyplot as plt
import pandas as pd
prets=pd.read_csv('/content/prets.csv')
v=prets.groupby('ville')['remboursement'].mean()
v=v.reset_index()
villes=v['ville']
for ville in villes:

  df=prets.loc[prets['ville']==ville,:]
  plt.scatter(df['revenu'], df['taux'], label=ville)
plt.legend()
villes


```

```python
import seaborn as sns
prets

sns.barplot(data=prets, x='ville', y='remboursement', ci=False, estimator=sum)
```

```python
plt.figure(figsize=(10,6))

plt.rcParams.update({'font.size': 10})

sns.scatterplot(data=prets, x='revenu', y='remboursement', hue='type')

plt.ylabel("remboursement")

plt.xlabel('Revenu mensuel (€)')

plt.grid()

plt.xlim(500, 7500)

plt.legend(bbox_to_anchor=(1, 1.02))

plt.title("Taux de remboursement en fonction du revenu mensuel\npar type de prêt contracté")

plt.show()
```
